# Mô tả dữ liệu

```python
import os

train_path='/content/drive/MyDrive/Dataset_bienbaoVN/Train'
test_path='/content/drive/MyDrive/Dataset_bienbaoVN/Test'

for folder in os.listdir(train_path):
  if len(folder)<3:
    new_folder=folder.zfill(3)
    os.rename(os.path.join(train_path,folder),os.path.join(train_path,new_folder))

for folder in os.listdir(test_path):
  if len(folder)<3:
    new_folder=folder.zfill(3)
    os.rename(os.path.join(test_path,folder),os.path.join(test_path,new_folder))
```

Mô tả tập train

```python
print('So luong folder trong train:',len(os.listdir(train_path)))

train_info=[]

total_train_images=0

for i in os.listdir(train_path):
  j=os.path.join(train_path,i)
  desc='Folder '+i+' : '+str(len(os.listdir(j)))+' images'
  train_info.append(desc)
  total_train_images+=len(os.listdir(j))

print('So luong anh trong train:',total_train_images)

train_info.sort()
train_info
```

Mô tả tập test

```python
test_info=[]

print('So luong folder trong test:',len(os.listdir(test_path)))

total_test_images=0

for i in os.listdir(test_path):
  j=os.path.join(test_path,i)
  desc='Folder '+i+' : '+str(len(os.listdir(j)))+' images'
  test_info.append(desc)
  total_test_images+=len(os.listdir(j))

print('So luong anh trong test:',total_test_images)

test_info.sort()
test_info
```

Mô tả các tên biển báo

```python
import pandas as pd

trafic_sign_path = '/content/drive/MyDrive/Dataset_bienbaoVN/class.csv'
traffic_sign_key = pd.read_csv(trafic_sign_path)

traffic_sign=[]
for i in traffic_sign_key:
  traffic_sign.append(i.strip())

traffic_sign
```

```python
import matplotlib.pyplot as plt

for j,k in enumerate(os.listdir(os.path.join(train_path,'000'))):
  if j<9:
    img=plt.imread(os.path.join(train_path,'000',k))
    plt.subplot(3,3,j+1)
    plt.imshow(img)
    plt.title(traffic_sign[0])
  else:
    break

plt.tight_layout()
plt.show()
```

# Khởi tạo dữ liệu train và test

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rescale=1./255,rotation_range=40,
                             width_shift_range=0.2,height_shift_range=0.2,
                             shear_range=0.2,zoom_range=0.2,brightness_range=[0.5,1.5],
                             fill_mode='nearest',validation_split=0.2)
```

```python

train_data=datagen.flow_from_directory(train_path,target_size=(64,64),color_mode='rgb',
                                        batch_size=32,class_mode='categorical',
                                       seed=42,shuffle=True,subset='training')

validation_data=datagen.flow_from_directory(train_path,target_size=(64,64),color_mode='rgb',
                                        batch_size=32,class_mode='categorical',
                                            seed=42,shuffle=True,subset='validation')

test_data=datagen.flow_from_directory(test_path,target_size=(64,64),color_mode='rgb',
                                        batch_size=32,class_mode='categorical',seed=42,shuffle=True)
```

VD: Hình ảnh trong tập train sau khi được khởi tạo

```python
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(train_data[0][0][i])
  plt.title(traffic_sign[np.argmax(train_data[0][1][i])])
plt.tight_layout()
plt.show()
```

VD: Hình ảnh của tập test sau khi khởi tạo

```python
for i in range(3):
  plt.subplot(1,3,i+1)
  plt.imshow(test_data[0][0][i])
  plt.title(traffic_sign[np.argmax(test_data[0][1][i])])
plt.tight_layout()
```

VD: Hình ảnh của tập validation sau khi khởi tạo

```python
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(validation_data[0][0][i])
  plt.title(traffic_sign[np.argmax(validation_data[0][1][i])])
plt.tight_layout()
plt.show()
```

# Xây dựng Model

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
```

Model 1: Dùng activation relu

```python
model=Sequential()

model.add(Conv2D(32,(3,3),activation='relu',padding='same',input_shape=(64,64,3)))
model.add(BatchNormalization())
model.add(Conv2D(32,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.2))

model.add(Conv2D(64,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(64,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.2))

model.add(Conv2D(128,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(128,(3,3),activation='relu',padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(121,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_model=model.fit(train_data,epochs=35,batch_size=64,validation_data=validation_data)
```

Đánh giá model1 - dùng activation: relu

```python
import matplotlib.pyplot as plt

train_acc=train_model.history['accuracy']
val_acc=train_model.history['val_accuracy']
train_loss=train_model.history['loss']
val_loss=train_model.history['val_loss']
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(train_acc,label='train_acc')
plt.plot(val_acc,label='val_acc')
plt.title('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(train_loss,label='train_loss')
plt.plot(val_loss,label='val_loss')
plt.legend()
plt.title('Loss')
plt.show()
```

Dự đoán hình ảnh biển báo của model 1 - dùng activation: relu

```python
test_model=model.predict(test_data[0][0])
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(test_data[0][0][0])
plt.title(traffic_sign[np.argmax(test_data[0][1][0])])
print('Dự đoán hình 1:', traffic_sign[np.argmax(test_model[0])])

plt.subplot(1,2,2)
plt.imshow(test_data[0][0][1])
plt.title(traffic_sign[np.argmax(test_data[0][1][1])])
print('Dự đoán hình 2:', traffic_sign[np.argmax(test_model[1])])
```

```python
model.save('/content/drive/MyDrive/colab/model_datagen3_bienbaoVN.h5')
```

Model 2: Dùng activation elu

```python
from keras.layers import ELU

model2=Sequential()

model2.add(Conv2D(32,(3,3),activation='elu',padding='same',input_shape=(64,64,3)))
model2.add(BatchNormalization())
model2.add(Conv2D(32,(3,3),activation='elu',padding='same'))
model2.add(BatchNormalization())
model2.add(MaxPooling2D(2,2))
model2.add(Dropout(0.2))

model2.add(Conv2D(64,(3,3),activation='elu',padding='same'))
model2.add(BatchNormalization())
model2.add(Conv2D(64,(3,3),activation='elu',padding='same'))
model2.add(BatchNormalization())
model2.add(MaxPooling2D(2,2))
model2.add(Dropout(0.2))

model2.add(Conv2D(128,(3,3),activation='elu',padding='same'))
model2.add(BatchNormalization())
model2.add(Conv2D(128,(3,3),activation='elu',padding='same'))
model2.add(BatchNormalization())
model2.add(MaxPooling2D(2,2))
model2.add(Dropout(0.2))

model2.add(Flatten())
model2.add(Dense(256,activation='elu'))
model2.add(BatchNormalization())
model2.add(Dense(121,activation='softmax'))

model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model2.summary()
```

```python
from keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='val_loss',patience=8,restore_best_weights=True)

train_model2=model2.fit(train_data,epochs=35,batch_size=64,validation_data=validation_data,
                        callbacks=[early_stop])
```

Đánh giá model2 - dùng activation elu

```python
import matplotlib.pyplot as plt
train_acc2=train_model2.history['accuracy']
val_acc2=train_model2.history['val_accuracy']
train_loss2=train_model2.history['loss']
val_loss2=train_model2.history['val_loss']
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(train_acc2,label='train_acc')
plt.plot(val_acc2,label='val_acc')
plt.title('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(train_loss2,label='train_loss')
plt.plot(val_loss2,label='val_loss')
plt.legend()
plt.title('Loss')
plt.show()
```

Dự đoán hình ảnh biển báo của model2 - dùng activation: elu

```python
test_model2=model2.predict(test_data[0][0])
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(test_data[0][0][0])
plt.title(traffic_sign[np.argmax(test_data[0][1][0])])
print('Dự đoán hình 1:', traffic_sign[np.argmax(test_model2[0])])

plt.subplot(1,2,2)
plt.imshow(test_data[0][0][1])
plt.title(traffic_sign[np.argmax(test_data[0][1][1])])
print('Dự đoán hình 2:', traffic_sign[np.argmax(test_model2[1])])
```

Model 3: Dùng kernel_initializer: he_uniform

```python
model3=Sequential()

model3.add(Conv2D(32,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same',input_shape=(64,64,3)))
model3.add(BatchNormalization())
model3.add(Conv2D(32,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model3.add(BatchNormalization())
model3.add(MaxPooling2D(2,2))
model3.add(Dropout(0.2))

model3.add(Conv2D(64,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model3.add(BatchNormalization())
model3.add(Conv2D(64,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model3.add(BatchNormalization())
model3.add(MaxPooling2D(2,2))
model3.add(Dropout(0.2))

model3.add(Conv2D(128,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model3.add(BatchNormalization())
model3.add(Conv2D(128,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model3.add(BatchNormalization())
model3.add(MaxPooling2D(2,2))
model3.add(Dropout(0.2))

model3.add(Flatten())
model3.add(Dense(256,activation='relu',kernel_initializer='he_uniform'))
model3.add(BatchNormalization())
model3.add(Dense(121,activation='softmax'))

model3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model3.summary()
```

```python
train_model3=model3.fit(train_data,epochs=35,batch_size=64,validation_data=validation_data)
```

Đánh giá model3 - dùng kernel_initializer: he_uniform

```python
import matplotlib.pyplot as plt

train_acc3=train_model3.history['accuracy']
val_acc3=train_model3.history['val_accuracy']
train_loss3=train_model3.history['loss']
val_loss3=train_model3.history['val_loss']
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(train_acc3,label='train_acc')
plt.plot(val_acc3,label='val_acc')
plt.title('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(train_loss3,label='train_loss')
plt.plot(val_loss3,label='val_loss')
plt.legend()
plt.title('Loss')
plt.show()
```

Dự đoán hình ảnh biển báo của model3 - dùng kernel_initializer: he_uniform

```python
test_model3=model3.predict(test_data[0][0])
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(test_data[0][0][0])
plt.title(traffic_sign[np.argmax(test_data[0][1][0])])
print('Dự đoán hình 1:', traffic_sign[np.argmax(test_model3[0])])

plt.subplot(1,2,2)
plt.imshow(test_data[0][0][1])
plt.title(traffic_sign[np.argmax(test_data[0][1][1])])
print('Dự đoán hình 2:', traffic_sign[np.argmax(test_model3[1])])

plt.show()
```
